```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# read file data (csv)
url = "https://raw.githubusercontent.com/keishapq31/Prak-Pengkom-05/main/H05.csv"
df = pd.read_csv(url)
```

Program di atas untuk mengimport library yang dibutuhkan dalam program ini secara keseluruhan, serta untuk membaca data yang digunakan.

```python
# 1. Histogram distribusi nilai Matematika
I = [20,30,40,50,60,70,80,90,100]

plt.hist(df["nilai_mat"],bins=I,rwidth=0.8,color="r")
plt.style.use("seaborn")
plt.title("Frekuensi nilai Matematika")
plt.xlabel("Nilai Matematika")
plt.ylabel("Frekuensi")
plt.figure(figsize=(16,6))

plt.show()
```

```python
# 2. Diagram batang horizontal jumlah mahasiswa masing-masing fakultas
df["fakultas"].value_counts().plot(kind="barh")
plt.style.use("seaborn")
plt.title("Jumlah mahasiswa tiap Fakultas/Sekolah")
```

```python
# 3. Diagram pie jumlah mahasiswa masing-masing fakultas
df["fakultas"].value_counts().plot(kind="pie")
```

4. Berdasarkan diagaram pie dan batang horizontal, manakah fakultas dengan mahasiswa terbanyak? Manakah diagram yang lebih baik dalam menampilkan fakultas terbanyak dan mengapa?
###### Saya pribadi merasa bahwa diagam batang horizontal dapat menunjukkan dengan lebih jelas perkiraan jumlahnya dalam angka serta perbandingannya dengan jumlah mahasiswa Fakultas/Sekolah lainnya. Namun diagram pie juga saya rasa bagus untuk menampilkan perbandingan dalam ukuran bagian, namun kurang jelas dalam menunjukkan berupa jumlah.

```python
# 5. Scatter plot dengan nilai Kimia sebagai x dan Fisika sebagai y
x = df["nilai_kim"]
y = df["nilai_fis"]
plt.figure(figsize=(16,6))
plt.style.use("seaborn")
plt.plot(x,y,"o",ms=2,c="blue")
plt.title("Scatter plot nilai Kimia dan Fisika")
plt.xlabel("Kimia")
plt.ylabel("Fisika")
plt.show()
```
